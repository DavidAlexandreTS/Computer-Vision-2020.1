```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img2 = cv.imread('fenix.jpg', 0)
img1 = cv.imread('phoenix.jpg', 0)
```

```python
plt.subplot(121), plt.imshow(img1), plt.title('Kikis Queens')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img2), plt.title('Chihiro and Haku')
plt.xticks([]), plt.yticks([])
plt.gcf().set_size_inches(16, 8)
plt.show()
```
